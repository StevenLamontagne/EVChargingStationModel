```python
import json
import matplotlib.pyplot as plt
import os 
import numpy as np

os.chdir("C:\\Users\\dobby\\Desktop\\Git Hub repo\\EVChargingStationModel\\Data\\GrowthFunctions")
```

```python
data = json.load(open("GrowthFunction_HomeCharging_FixedSolution.txt"))
```

```python
slope = np.array([data[s]["Slope"] for s in data])
intercept = np.array([data[s]["Intercept"] for s in data])
```

```python
x = [data[s]["Startpoint"] for s in data]
y = list(slope*np.array(x) + intercept)
```

```python
x.append(1)
y.append(1)
```

```python
fig, ax = plt.subplots(1,1)

ax.plot(x[:4],y[:4], 's-r')
ax.plot(x[4:],y[4:], '--b')
# ax.set_xlabel("$\phi (t)$")
# ax.set_ylabel("$\phi (t+1)$")
ax.set_xlabel("Fraction of the Population with EVs in year $t$")
ax.set_ylabel("Fraction of the Population with EVs in year $t+1$")
ax.legend(["Aggregated", "Extended"])
```
